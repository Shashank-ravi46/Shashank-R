```python
# CELL 11
# Pull the most important line items into a small analysis table

def safe_extract_row(df, keywords):
    """
    Find the first row whose index contains any of the keywords.
    """
    if df.empty:
        return None

    idx = df.index.astype(str).str.lower()
    for keyword in keywords:
        matches = idx.str.contains(keyword, na=False)
        if matches.any():
            return df.loc[matches].iloc[0]
    return None


revenue_row = safe_extract_row(income_clean_df, ["revenue", "sales"])
net_income_row = safe_extract_row(income_clean_df, ["net income", "profit"])
gross_profit_row = safe_extract_row(income_clean_df, ["gross profit"])

summary_rows = []

if revenue_row is not None:
    summary_rows.append(("Revenue", revenue_row))
if gross_profit_row is not None:
    summary_rows.append(("Gross Profit", gross_profit_row))
if net_income_row is not None:
    summary_rows.append(("Net Income", net_income_row))

summary_table = pd.DataFrame({name: row for name, row in summary_rows}).T

print("Summary table created.")
display(summary_table)

```

```python
# CELL 19
# Create simple IBM forecast assumptions

forecast_years = 5

revenue_growth_rates = [0.04, 0.035, 0.03, 0.03, 0.025]
operating_margin_assumptions = [0.17, 0.175, 0.18, 0.182, 0.185]
net_margin_assumptions = [0.12, 0.123, 0.125, 0.127, 0.13]
fcf_margin_assumptions = [0.13, 0.135, 0.14, 0.142, 0.145]

wacc = 0.09
terminal_growth = 0.025

print("Forecast assumptions created.")
```

```python
# CELL 20
# Build a simple 5-year forecast table

# Get the latest historical revenue
# To fix the AttributeError: 'NoneType' object has no attribute 'iloc'
# This error occurs because the `revenue` variable from the previous cell (CELL 14) is `None`.
# This happened because the `find_row_by_keywords` function in CELL 14 searched the DataFrame's *index* for keywords,
# but the `income_clean_df` (from which `revenue` was supposed to be extracted) has its line item names
# (e.g., 'us-gaap_Revenues') in a 'concept' column, not its numerical index.

# We need to re-extract the latest historical revenue correctly from `income_clean_df` within this cell.

latest_revenue = 0.0 # Default value in case extraction fails

if not income_clean_df.empty and 'concept' in income_clean_df.columns:
    concept_col_lower = income_clean_df['concept'].astype(str).str.lower()

    # Find the row where the 'concept' column contains "revenues"
    revenue_row_matches = concept_col_lower.str.contains("revenues", na=False)

    if revenue_row_matches.any():
        revenue_row_series = income_clean_df.loc[revenue_row_matches].iloc[0]

        # Identify columns that are not 'concept' and likely contain financial data
        financial_data_columns = [col for col in income_clean_df.columns if col != 'concept']

        # Iterate through financial columns to find the latest revenue value.
        # Assuming the latest year is in the first column of financial data displayed.
        for col in financial_data_columns:
            if col in revenue_row_series.index: # Check if column exists in the Series
                value_str = str(revenue_row_series[col]).replace('$', '').replace(',', '')
                try:
                    current_parsed_value = float(value_str)
                    # Found the latest revenue value, break the loop
                    latest_revenue = current_parsed_value
                    break
                except ValueError:
                    # Not a valid float, continue to the next column
                    continue

    if latest_revenue == 0.0:
        print("Warning: Could not extract latest numerical revenue from `income_clean_df` using 'revenues' keyword. Using 0.0 for forecast.")
else:
    print("Warning: `income_clean_df` is empty or missing 'concept' column. Using 0.0 for forecast.")

# Ensure latest_revenue is float for calculations
latest_revenue = float(latest_revenue)

forecast_data = []
current_revenue = latest_revenue

for i in range(forecast_years):
    year_label = f"Year {i+1}"

    growth = revenue_growth_rates[i]
    op_margin = operating_margin_assumptions[i]
    net_margin = net_margin_assumptions[i]
    fcf_margin = fcf_margin_assumptions[i]

    current_revenue = current_revenue * (1 + growth)
    operating_income_forecast = current_revenue * op_margin
    net_income_forecast = current_revenue * net_margin
    free_cash_flow_forecast = current_revenue * fcf_margin

    forecast_data.append({
        "Year": year_label,
        "Revenue": current_revenue,
        "Revenue Growth": growth,
        "Operating Margin": op_margin,
        "Operating Income": operating_income_forecast,
        "Net Margin": net_margin,
        "Net Income": net_income_forecast,
        "FCF Margin": fcf_margin,
        "Free Cash Flow": free_cash_flow_forecast
    })

forecast_df = pd.DataFrame(forecast_data)
display(forecast_df)
```

```python
# CELL 21
# Calculate discounted cash flow valuation

fcf_values = forecast_df["Free Cash Flow"].values

discounted_fcfs = []
for t, fcf in enumerate(fcf_values, start=1):
    pv = fcf / ((1 + wacc) ** t)
    discounted_fcfs.append(pv)

forecast_df["Discounted FCF"] = discounted_fcfs

terminal_fcf = fcf_values[-1] * (1 + terminal_growth)
terminal_value = terminal_fcf / (wacc - terminal_growth)
discounted_terminal_value = terminal_value / ((1 + wacc) ** forecast_years)

enterprise_value = forecast_df["Discounted FCF"].sum() + discounted_terminal_value

print("Discounted forecast table:")
display(forecast_df)

print("\nTerminal Value:", terminal_value)
print("Discounted Terminal Value:", discounted_terminal_value)
print("Enterprise Value:", enterprise_value)
```

```python
# CELL 22
# Historical vs forecast revenue chart

# Re-extract historical revenue from income_clean_df to fix AttributeError
historical_revenue_data = {}
if not income_clean_df.empty and 'concept' in income_clean_df.columns:
    # Find the row corresponding to 'Revenues'
    revenue_row_df = income_clean_df[income_clean_df['concept'].str.contains('revenues', case=False, na=False)]

    if not revenue_row_df.empty:
        # Get columns that represent financial years (e.g., '2023-12-31 (FY)')
        financial_year_columns = [col for col in revenue_row_df.columns if 'FY' in col and '-' in col]

        # Sort columns to ensure they are in chronological order for plotting
        financial_year_columns_sorted = sorted(financial_year_columns, key=lambda x: int(x.split('-')[0]))

        # Extract values and use just the year for the period label
        for col in financial_year_columns_sorted:
            year_label = col.split('-')[0] # e.g., '2023' from '2023-12-31 (FY)'
            value_str = str(revenue_row_df[col].iloc[0]).replace('$', '').replace(',', '')
            try:
                historical_revenue_data[year_label] = float(value_str)
            except ValueError:
                # Skip if value cannot be converted to float
                continue

# Create a Series for historical revenue
historical_revenue_series = pd.Series(historical_revenue_data)

# Prepare historical_chart_df
historical_chart_df = pd.DataFrame({
    "Period": historical_revenue_series.index.astype(str), # Ensure periods are strings
    "Revenue": historical_revenue_series.values,
    "Type": "Historical"
})

# Prepare forecast_chart_df
forecast_revenue = forecast_df[["Year", "Revenue"]].copy()
forecast_revenue.columns = ["Period", "Revenue"]

forecast_chart_df = pd.DataFrame({
    "Period": forecast_revenue["Period"],
    "Revenue": forecast_revenue["Revenue"],
    "Type": "Forecast"
})

# Combine and plot
combined_chart_df = pd.concat([historical_chart_df, forecast_chart_df], ignore_index=True)

fig = px.line(
    combined_chart_df,
    x="Period",
    y="Revenue",
    color="Type",
    markers=True,
    title="IBM Historical and Forecast Revenue"
)

fig.update_layout(template="plotly_white")
fig.show()
```

```python
import pandas as pd
import plotly.express as px

def get_revenue_series(df_financials, company_name):
    """Extracts historical revenue series from a cleaned financial DataFrame."""
    revenue_data = {}
    if df_financials.empty or 'concept' not in df_financials.columns:
        return pd.Series(revenue_data)

    # Filter for rows where the 'concept' column contains 'revenue'
    # This captures 'us-gaap_Revenues', 'us-gaap_RevenueNet', etc.
    revenue_rows = df_financials[df_financials['concept'].str.contains('revenue', case=False, na=False)]

    if revenue_rows.empty:
        return pd.Series(revenue_data)

    # Iterate through the rows that are identified as revenue
    for _, row in revenue_rows.iterrows():
        # Look for columns that represent fiscal years, typically ending with '(FY)'
        financial_year_cols = [col for col in row.index if 'FY' in col and '-' in col]

        # If a row has multiple financial year columns (like IBM's structure)
        if len(financial_year_cols) > 1:
            for col in financial_year_cols:
                year_label = col.split('-')[0]  # e.g., '2023' from '2023-12-31 (FY)'
                value_str = str(row[col]).replace('$', '').replace(',', '')
                try:
                    value = float(value_str)
                    revenue_data[year_label] = value
                except ValueError:
                    continue
        # If a row has only one financial year column (like PLTR's structure with repeated revenue rows)
        elif len(financial_year_cols) == 1:
            col = financial_year_cols[0]
            year_label = col.split('-')[0]
            value_str = str(row[col]).replace('$', '').replace(',', '')
            try:
                value = float(value_str)
                revenue_data[year_label] = value
            except ValueError:
                continue

    # Sort by year to ensure correct plotting order
    if revenue_data:
        sorted_years = sorted(revenue_data.keys())
        sorted_revenue_data = {year: revenue_data[year] for year in sorted_years}
        return pd.Series(sorted_revenue_data)
    else:
        return pd.Series(revenue_data)

# CELL 32
# Build revenue comparison chart data

# Get IBM historical revenue
ibm_historical_revenue_series = get_revenue_series(income_clean_df, "IBM")

ibm_revenue_df = pd.DataFrame({
    "Period": ibm_historical_revenue_series.index.astype(str),
    "Revenue": ibm_historical_revenue_series.values,
    "Company": "IBM"
})

# Get PLTR historical revenue
pltr_historical_revenue_series = get_revenue_series(pltr_data["income_df"], "PLTR")

pltr_revenue_df = pd.DataFrame({
    "Period": pltr_historical_revenue_series.index.astype(str),
    "Revenue": pltr_historical_revenue_series.values,
    "Company": "PLTR"
})

revenue_comparison_df = pd.concat([ibm_revenue_df, pltr_revenue_df], ignore_index=True)

fig = px.line(
    revenue_comparison_df,
    x="Period",
    y="Revenue",
    color="Company",
    markers=True,
    title="IBM vs PLTR Historical Revenue"
)

fig.update_layout(template="plotly_white")
fig.show()
```

```python
# CELL 34
# Palantir forecast assumptions

pltr_forecast_years = 5

pltr_revenue_growth_rates = [0.24, 0.20, 0.17, 0.14, 0.12]
pltr_operating_margin_assumptions = [0.18, 0.21, 0.24, 0.26, 0.28]
pltr_net_margin_assumptions = [0.16, 0.18, 0.20, 0.22, 0.23]
pltr_fcf_margin_assumptions = [0.22, 0.24, 0.25, 0.26, 0.27]

pltr_wacc = 0.10
pltr_terminal_growth = 0.03

print("Palantir forecast assumptions created.")
```

```python
# CELL 35
# Build Palantir forecast table

# To fix the AttributeError: 'NoneType' object has no attribute 'iloc',
# we need to correctly extract the latest historical revenue for PLTR.
# The `pltr_revenue` variable from CELL 27 is None because `find_row_by_keywords`
# incorrectly searches the index instead of the 'concept' column.
# We will reuse the `get_revenue_series` function defined in CELL 32.

pltr_historical_revenue_series = get_revenue_series(pltr_data["income_df"], "PLTR")

pltr_latest_revenue = 0.0 # Default value
if not pltr_historical_revenue_series.empty:
    pltr_latest_revenue = pltr_historical_revenue_series.iloc[-1]
else:
    print("Warning: Could not extract latest numerical revenue for PLTR. Using 0.0 for forecast.")

pltr_forecast_data = []
pltr_current_revenue = pltr_latest_revenue

for i in range(pltr_forecast_years):
    year_label = f"Year {i+1}"

    growth = pltr_revenue_growth_rates[i]
    op_margin = pltr_operating_margin_assumptions[i]
    net_margin = pltr_net_margin_assumptions[i]
    fcf_margin = pltr_fcf_margin_assumptions[i]

    pltr_current_revenue = pltr_current_revenue * (1 + growth)
    pltr_operating_income_forecast = pltr_current_revenue * op_margin
    pltr_net_income_forecast = pltr_current_revenue * net_margin
    pltr_free_cash_flow_forecast = pltr_current_revenue * fcf_margin

    pltr_forecast_data.append({
        "Year": year_label,
        "Revenue": pltr_current_revenue,
        "Revenue Growth": growth,
        "Operating Margin": op_margin,
        "Operating Income": pltr_operating_income_forecast,
        "Net Margin": net_margin,
        "Net Income": pltr_net_income_forecast,
        "FCF Margin": fcf_margin,
        "Free Cash Flow": pltr_free_cash_flow_forecast
    })

pltr_forecast_df = pd.DataFrame(pltr_forecast_data)
display(pltr_forecast_df)
```

```python
# CELL 36
# Calculate Palantir DCF

pltr_fcf_values = pltr_forecast_df["Free Cash Flow"].values

pltr_discounted_fcfs = []
for t, fcf in enumerate(pltr_fcf_values, start=1):
    pv = fcf / ((1 + pltr_wacc) ** t)
    pltr_discounted_fcfs.append(pv)

pltr_forecast_df["Discounted FCF"] = pltr_discounted_fcfs

pltr_terminal_fcf = pltr_fcf_values[-1] * (1 + pltr_terminal_growth)
pltr_terminal_value = pltr_terminal_fcf / (pltr_wacc - pltr_terminal_growth)
pltr_discounted_terminal_value = pltr_terminal_value / ((1 + pltr_wacc) ** pltr_forecast_years)

pltr_enterprise_value = pltr_forecast_df["Discounted FCF"].sum() + pltr_discounted_terminal_value

print("Discounted forecast table:")
display(pltr_forecast_df)

print("\nTerminal Value:", pltr_terminal_value)
print("Discounted Terminal Value:", pltr_discounted_terminal_value)
print("Enterprise Value:", pltr_enterprise_value)
```

```python
# CELL 37
# Historical vs forecast PLTR revenue chart

# The previous error 'AttributeError: 'NoneType' object has no attribute 'copy'' occurred because
# `pltr_revenue` was None. This was due to the `find_row_by_keywords` function failing to
# correctly extract revenue data from pltr_data["income_df"] as it searched the index,
# not the 'concept' column.

# We will use the already defined `get_revenue_series` function (now improved in CELL 32) to properly extract PLTR's
# historical revenue.

pltr_historical_revenue_series = get_revenue_series(pltr_data["income_df"], "PLTR")

pltr_historical_chart_df = pd.DataFrame()
if not pltr_historical_revenue_series.empty:
    pltr_historical_chart_df = pd.DataFrame({
        "Period": pltr_historical_revenue_series.index.astype(str),
        "Revenue": pltr_historical_revenue_series.values,
        "Type": "Historical"
    })
else:
    print("Warning: No historical revenue data found for PLTR. Historical data will not be plotted.")


pltr_forecast_revenue_chart = pltr_forecast_df[["Year", "Revenue"]].copy()
pltr_forecast_revenue_chart.columns = ["Period", "Revenue"]

pltr_forecast_chart_df = pd.DataFrame({
    "Period": pltr_forecast_revenue_chart["Period"],
    "Revenue": pltr_forecast_revenue_chart["Revenue"],
    "Type": "Forecast"
})

# Combine historical and forecast data. Handle cases where historical_chart_df might be empty.
if not pltr_historical_chart_df.empty:
    pltr_combined_chart_df = pd.concat([pltr_historical_chart_df, pltr_forecast_chart_df], ignore_index=True)
else:
    # If no historical data, just use the forecast data for plotting
    pltr_combined_chart_df = pltr_forecast_chart_df.copy()

if not pltr_combined_chart_df.empty:
    fig = px.line(
        pltr_combined_chart_df,
        x="Period",
        y="Revenue",
        color="Type",
        markers=True,
        title="PLTR Historical and Forecast Revenue"
    )

    fig.update_layout(template="plotly_white")
    fig.show()
else:
    print("No data available to plot PLTR Historical and Forecast Revenue.")
```
